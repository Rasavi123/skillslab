# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
# SQL query to fetch the "ConvertedComp" column
sql_query = 'SELECT ConvertedComp FROM master;'

# Use pandas to execute the SQL query and fetch data into a DataFrame
df = pd.read_sql_query(sql_query, conn)

# Close the database connection
#conn.close()

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Converted Compensation (USD)')
plt.ylabel('Frequency')
plt.title('Histogram of Converted Compensation')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

sql_query = 'SELECT Age FROM master;'

# Use pandas to execute the SQL query and fetch data into a DataFrame
df = pd.read_sql_query(sql_query, conn)  
plt.figure(figsize=(8, 6)) 
df['Age'].plot(kind='box')
#plt.xlabel('Age')
plt.title('Box Plot of Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
sql_query = 'SELECT Age, WorkWeekHrs FROM master;'

# Use pandas to execute the SQL query and fetch data into a DataFrame
df = pd.read_sql_query(sql_query, conn)
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], color='skyblue', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
# SQL query to fetch the "WorkWeekHrs", "CodeRevHrs", and "Age" columns
sql_query = 'SELECT WorkWeekHrs, CodeRevHrs, Age FROM master;'

# Use pandas to execute the SQL query and fetch data into a DataFrame
df = pd.read_sql_query(sql_query, conn)

plt.figure(figsize=(10, 6))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age']*5, alpha=0.7, cmap='viridis')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs (Bubble Size: Age)')
plt.colorbar(label='Age')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

# SQL query to fetch the top 5 databases that respondents wish to learn next year and their counts
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df_db = pd.read_sql_query(QUERY,conn)
df_count=df_db['DatabaseDesireNextYear'].value_counts().rename_axis('DatabaseName').reset_index(name='Counts')
df_top5=df_count[:5].copy()

plt.pie(df_top5['Counts'],labels=df_top5['DatabaseName'],autopct='%.2f')
plt.title('Top 5 databases to Learn')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """
SELECT Age,WorkWeekHrs,CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df_wk = pd.read_sql_query(QUERY,conn)

df_g=df_wk.groupby('Age').median()

df_g.plot(kind='bar',stacked=True)
plt.title('Median workweek hrs and CodeRevHrs in 30-35 age group')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

sql_query = '''
    SELECT Age,ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60;
'''

# Use pandas to execute the SQL query and fetch data into a DataFrame
df = pd.read_sql_query(sql_query, conn)
new_df=df.groupby('Age').median()

plt.figure(figsize=(10, 6))
new_df.plot(kind='line', marker='o', color='skyblue')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

sql_query = '''
    SELECT MainBranch, COUNT(*) AS Count
    FROM master
    GROUP BY MainBranch
    ORDER BY Count DESC;
'''

# Use pandas to execute the SQL query and fetch data into a DataFrame
df = pd.read_sql_query(sql_query, conn)

plt.figure(figsize=(8, 4))
plt.barh(df['MainBranch'], df['Count'], color='skyblue')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.title('MainBranch Distribution')
plt.grid(axis='x')
plt.show()
```
